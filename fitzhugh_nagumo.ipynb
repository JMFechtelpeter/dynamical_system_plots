```python
from continuous_DS import PhasePlanePlot

a = -1
b = 1/2
I = 1
tau = 10
initial = [0,1.00002]
Tmax = 1000

fitzhugh = lambda v, w: (v - (1/3)*v**3 - w + I, (1/tau)*(v - a - b*w))

v_nullcline = lambda v, I=I: v - (1/3)*v**3 + I
w_nullcline = lambda v, a=a, b=b: (v - a)/b

ppplot = PhasePlanePlot(fitzhugh, vectorized=False, xlim=(-3,3), ylim=(-3,3))
_, traj = ppplot.trajectory(Tmax, initial)
ppplot.plot_flow()
ppplot.plot_axes(c='k')
ppplot.plot(traj[0], traj[1])
ppplot.plot_curve(v_nullcline)
ppplot.plot_curve(w_nullcline)
```
