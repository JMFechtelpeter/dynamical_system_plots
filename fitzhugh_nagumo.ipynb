```python
from continuous_DS import DynamicalSystem2D

a = -1
b = 1/2
I = 1
tau = 10
x0 = 0
y0 = 1.00002
Tmax = 1000

vdot = lambda v, w, I=I, **params: v - (1/3)*v**3 - w + I
wdot = lambda v, w, a=a, b=b, tau=tau, **params: (1/tau)*(v - a - b*w)

v_nullcline = lambda v, I=I, **params: v - (1/3)*v**3 + I
w_nullcline = lambda v, a=a, b=b, **params: (v - a)/b

system = DynamicalSystem2D(vdot, wdot, (-3,3), (-3,3))
_, traj = system.trajectory(Tmax, x0, y0, I=I, tau=tau)
ax = system.plot_axes(c='k')
ax = system.plot_trajectory(traj[0], traj[1], ax=ax)
ax = system.plot_curve(v_nullcline, ax=ax)
ax = system.plot_curve(w_nullcline, ax=ax)
```
